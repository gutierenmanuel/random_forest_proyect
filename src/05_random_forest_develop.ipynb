```python
import pandas as pd

# load the csv data sets

# Con Index = False, no guardamos el indice en el archivo resultante
X_train = pd.read_csv("../data/processed/X_train.csv")
X_test = pd.read_csv("../data/processed/X_test.csv")

y_train = pd.read_csv("../data/processed/y_train.csv")
y_test = pd.read_csv("../data/processed/y_test.csv")
```

```python

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 100, random_state = 42)
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)
y_pred_train
```

```python
y_pred_test = model.predict(X_test)
y_pred_test
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import numpy as np

real = y_train
prediction = y_pred_train

# Calcular las métricas de evaluación
accuracy = accuracy_score(real, prediction)
precision = precision_score(real, prediction)
recall = recall_score(real, prediction)
f1 = f1_score(real, prediction)

# Mostrar los resultados
print(f'Accuracy: {accuracy:.8f}')
print(f'Precision: {precision:.8f}')
print(f'Recall: {recall:.8f}')
print(f'F1 Score: {f1:.8f}')
```

```python
from sklearn.tree import export_graphviz
import graphviz

for i in range(10):
    # Selecciona uno de los árboles (puedes elegir el índice que desees)
    tree_number = i
    tree = model.estimators_[tree_number]

    # Exporta el árbol a un archivo .dot
    dot_data = export_graphviz(tree, out_file=None, filled=True, rounded=True, special_characters=True)

    # Visualiza el árbol usando Graphviz
    graph = graphviz.Source(dot_data)
    graph.render(f"../docs/random_forest_tree_{tree_number}")  # Esto crea un archivo PDF con el árbol
```

```python
from pdf2image import convert_from_path
import os

# Ruta de la carpeta que contiene los archivos PDF
ruta_carpeta = '../docs/'

# Obtén la lista de archivos en la carpeta
archivos_en_carpeta = os.listdir(ruta_carpeta)

# Filtra solo los archivos PDF
archivos_pdf = [archivo for archivo in archivos_en_carpeta if archivo.lower().endswith('.pdf')]

# Procesa cada archivo PDF
for archivo_pdf in archivos_pdf:
    ruta_pdf = os.path.join(ruta_carpeta, archivo_pdf)

    # Convierte el PDF a una lista de imágenes
    imagenes = convert_from_path(ruta_pdf)

    # Guarda cada imagen como un archivo PNG
    for i, imagen in enumerate(imagenes):
        nombre_archivo = f'{os.path.splitext(archivo_pdf)[0]}_pagina_{i + 1}.png'
        ruta_guardado = os.path.join(ruta_carpeta, nombre_archivo)
        imagen.save(ruta_guardado, 'PNG')
        print(f"Guardada la imagen: {ruta_guardado}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


# Crear el modelo Random Forest
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = rf_classifier.predict(X_test)

# Evaluar la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo Random Forest en el conjunto de prueba: {accuracy:.2f}")

# Obtener la importancia de las características
feature_importances = rf_classifier.feature_importances_

# Obtener nombres de características
feature_names = X_train.columns

# Crear un DataFrame para facilitar la visualización
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})

# Ordenar el DataFrame por importancia
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Graficar la importancia de las características
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Importance')
plt.title('Random Forest Feature Importance')
plt.show()
```

```python

#Optimizacion de hiperparametros
from sklearn.model_selection import GridSearchCV
import numpy as np

#Definimos los parametros a mano que queremos ajustar
hyperparams = {
    'bootstrap': [True],
    'ccp_alpha': [0.0, 0.0001, 0.001, 0.01, 0,1],
    'class_weight': [None, 'balanced', 'balanced_subsample'],
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': [None, 1],
    'max_features': ['sqrt', 'log2', None],
    'max_leaf_nodes': [None],
    'max_samples': [None],
    'min_impurity_decrease': [0.0],
    'min_samples_leaf': [1],
    'min_samples_split': [2],
    'min_weight_fraction_leaf': [0.0],
    'n_estimators': [100],
    'n_jobs': [-2],
    'oob_score': [True],
    'random_state': [42],
    'verbose': [0],
    'warm_start': [True]
 }

# Inicializamos el grid
grid = GridSearchCV(model, hyperparams, scoring = "accuracy", cv = 4, n_jobs=-2)
grid
```

```python
# Ocultamos los warnings para mantener limpio el codigo
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

#Entrenamos el GridSearchCV
grid.fit(X_train, y_train)

print(f"Best hyperparameters: {grid.best_params_}")
```

```python
rf_classifier = RandomForestClassifier(**grid.best_params_)
rf_classifier.fit(X_train, y_train)
```
